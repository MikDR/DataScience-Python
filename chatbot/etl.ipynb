```python
import pandas as pd
```

```python
file_path = 'csv/TMDB_movie_dataset_v11.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
elimino = ["backdrop_path", "homepage", "imdb_id", "poster_path", "tagline", "spoken_languages", "keywords"]
df.drop(columns = elimino, inplace = True)
```

```python
numero_di_righe = len(df)
print("Il dataframe ha", numero_di_righe, "righe.")

df.head()
```

```python
df = df.dropna(subset=['title'])
df = df.dropna(subset=['original_title'])
```

```python
numero_di_righe = len(df)
print("Il dataframe ha", numero_di_righe, "righe.")
```
